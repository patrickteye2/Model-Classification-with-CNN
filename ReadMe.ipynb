# Fashion MNIST Classification with CNN

This project demonstrates how to use a Convolutional Neural Network (CNN) to classify images of clothing from the Fashion MNIST dataset. The dataset contains 70,000 grayscale images of 10 different fashion categories, such as shirts, trousers, and shoes.

## Project Overview

- **Dataset:** Fashion MNIST (28x28 grayscale images, 10 classes)
- **Goal:** Build a CNN model to accurately classify clothing items
- **Tools:** Python, TensorFlow/Keras, Matplotlib

## Steps

1. **Data Loading & Preprocessing**
   - Load the Fashion MNIST dataset from Keras
   - Reshape and normalize the images
   - Convert labels to one-hot encoding

2. **Model Building**
   - Construct a CNN with multiple convolutional and dense layers
   - Use ReLU activation for hidden layers and softmax for output

3. **Training**
   - Compile the model with Adam optimizer and categorical cross-entropy loss
   - Train the model and validate on a split of the training data

4. **Evaluation**
   - Evaluate the model on the test set
   - Plot training/validation accuracy and loss

5. **Prediction**
   - Predict classes for random test images
   - Visualize predictions and compare with true labels

## Results

- The model achieves high accuracy (around 90%) on the test set.
- Training and validation curves indicate the model is neither overfitting nor underfitting.
- The model can correctly classify most images, but may confuse visually similar items (e.g., coats and shirts).

## How to Run

1. Install dependencies:
   ```
   pip install tensorflow matplotlib numpy
   ```
2. Open the Jupyter notebook (`python_script.ipynb`) and run all cells.

## Interpretation

- **Good performance:** High accuracy and low loss on both training and validation sets.
- **Predictions:** The model can identify most clothing items, but some errors are reasonable due to visual similarity in grayscale images.
